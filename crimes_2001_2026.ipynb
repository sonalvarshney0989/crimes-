```python
import pandas as pd

files = glob.glob(r"C:/Users/Admin Pc/Downloads/crimes data/crimes data/*.csv")
cols = [
    "ID", "Case Number", "Date", "Block", "Primary Type",
    "Description", "Location Description", "Arrest",
    "Domestic", "Beat", "District", "Ward",
    "Community Area", "Year", "Latitude", "Longitude"
]

df_list = []

for file in files:
    try:
        df = pd.read_csv(file, usecols=cols)
        df_list.append(df)
    except Exception as e:
        print(f"Error in file: {file}")
        print(e)


merge_df = pd.concat(df_list, ignore_index=True)

print(merge_df.shape)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df = df.sort_values(by='Date', ascending=False)

# 5 lakhs
df_sample = df.head(500000).copy()

print(df_sample.shape)
```

```python
# Missing values check
print(df_sample.isnull().sum())

# Date missing remove
df_sample = df_sample.dropna(subset=['Date'])

# Categorical columns fill
cat_cols = ['Primary Type', 'Description', 'Location Description']

for col in cat_cols:
    df_sample[col] = df_sample[col].fillna('Unknown')

# Numeric columns fill
num_cols = ['Beat', 'District', 'Ward', 'Community Area']

for col in num_cols:
    df_sample[col] = df_sample[col].fillna(-1)
```

```python
df_sample = df_sample.drop_duplicates(subset='ID')
```

```python
df_sample['Arrest'] = df_sample['Arrest'].astype(bool)
df_sample['Domestic'] = df_sample['Domestic'].astype(bool)
```

```python
#Time Features
df_sample['Hour'] = df_sample['Date'].dt.hour
df_sample['DayOfWeek'] = df_sample['Date'].dt.day_name()
df_sample['Month'] = df_sample['Date'].dt.month
df_sample['Year'] = df_sample['Date'].dt.year
```

```python
#✅ 6. Data Quality Check
#✔️ Date Range check
print(df_sample['Date'].min(), df_sample['Date'].max())
```

```python
#Latitude/Longitude remove
df_sample = df_sample[
    (df_sample['Latitude'].between(-90, 90)) &
    (df_sample['Longitude'].between(-180, 180))
]
```

```python
#   Feature Engineering
```

```python
#  1. Temporal Features
```

```python
# Hour, Day, Month
df['Hour'] = df['Date'].dt.hour
df['DayOfWeek'] = df['Date'].dt.dayofweek   # 0=Monday
df['Month'] = df['Date'].dt.month

# Weekend flag
df['IsWeekend'] = df['DayOfWeek'].isin([5, 6]).astype(int)
```

```python
# Season Feature
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df['Month'].apply(get_season)
```

```python
#  2. Geographic Features
```

```python
#  Coordinate Binning
```

```python
df['Lat_bin'] = pd.cut(df['Latitude'], bins=50, labels=False)
df['Lon_bin'] = pd.cut(df['Longitude'], bins=50, labels=False)
```

```python
# District Clustering (KMeans)
```

```python
from sklearn.cluster import KMeans

coords = df[['Latitude', 'Longitude']].dropna()

kmeans = KMeans(n_clusters=10, random_state=42)
df.loc[coords.index, 'GeoCluster'] = kmeans.fit_predict(coords)
```

```python
# 3. Crime Severity Score
```

```python
severity_map = {
    'THEFT': 2,
    'BATTERY': 3,
    'ASSAULT': 4,
    'ROBBERY': 5,
    'HOMICIDE': 10,
    'BURGLARY': 4,
    'CRIMINAL DAMAGE': 3
}

df['Severity'] = df['Primary Type'].map(severity_map)
df['Severity'] = df['Severity'].fillna(1)
```

```python
# 4. Categorical Encoding
# Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder

le_crime = LabelEncoder()
df['CrimeEncoded'] = le_crime.fit_transform(df['Primary Type'])

le_loc = LabelEncoder()
df['LocationEncoded'] = le_loc.fit_transform(df['Location Description'])
```

```python
# One-Hot Encoding (optional)
```

```python
df = pd.get_dummies(df, columns=['Season'], drop_first=True)
```

```python
# 5. Normalize Coordinates
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['Latitude_norm', 'Longitude_norm']] = scaler.fit_transform(
    df[['Latitude', 'Longitude']]
)
```

```python
# Bonus Features
```

```python
#  Crime Density Feature
```

```python

df['AreaCrimeCount'] = df.groupby(['Lat_bin', 'Lon_bin'])['ID'].transform('count')
```

```python
# Time-Based Risk Feature
```

```python
df['HourlyCrimeCount'] = df.groupby('Hour')['ID'].transform('count')
```

```python
# Elbow Method (Optimal Clusters)
```

```python
import matplotlib.pyplot as plt

inertia = []

for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(coords)
    inertia.append(km.inertia_)

plt.plot(range(2, 11), inertia)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
# 3. DBSCAN (Density-Based Clustering)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.02, min_samples=10)
df.loc[coords.index, 'DBSCAN_Cluster'] = dbscan.fit_predict(coords)
```

```python
#4. Hierarchical Clustering
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

linked = linkage(coords.sample(5000), method='ward')  

plt.figure(figsize=(10,5))
dendrogram(linked)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()
```

```python
                                         # Unsupervised Learning - Dimensionality Reduction
```

```python
#1. Feature Selection 
```

```python
features = [
    'Hour', 'DayOfWeek', 'Month', 'IsWeekend',
    'Latitude_norm', 'Longitude_norm',
    'Severity', 'AreaCrimeCount', 'HourlyCrimeCount',
    'CrimeEncoded', 'LocationEncoded'
]

X = df[features].dropna()
```

```python
# 2. Feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 3. PCA (Principal Component Analysis)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Explained Variance
```

```python
print(pca.explained_variance_ratio_)
print("Total Variance:", sum(pca.explained_variance_ratio_))
```

```python
# Scree Plot
```

```python
import matplotlib.pyplot as plt

plt.plot(pca.explained_variance_ratio_, marker='o')
plt.title("PCA Explained Variance")
plt.xlabel("Principal Components")
plt.ylabel("Variance Ratio")
plt.show()
```

```python
# 4. Top Important Features
```

```python
import numpy as np

loadings = pd.DataFrame(
    pca.components_.T,
    columns=['PC1', 'PC2', 'PC3'],
    index=features
)

# Absolute importance
loadings['Importance'] = np.abs(loadings).sum(axis=1)

top_features = loadings.sort_values(by='Importance', ascending=False)

print(top_features.head(5))
```

```python
# K-Means Experiment Tracking
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

with mlflow.start_run(run_name="KMeans_Clustering"):

    k = 7
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)

    sil_score = silhouette_score(X_scaled, labels)

    #  Parameters log 
    mlflow.log_param("algorithm", "KMeans")
    mlflow.log_param("n_clusters", k)

    #  Metrics log
    mlflow.log_metric("silhouette_score", sil_score)

    print("Silhouette Score:", sil_score)
```

```python
# DBSCAN Tracking
```

```python
from sklearn.cluster import DBSCAN

with mlflow.start_run(run_name="DBSCAN_Clustering"):

    eps = 0.02
    min_samples = 10

    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(X_scaled)

    # Noise 
    mask = labels != -1

    if len(set(labels[mask])) > 1:
        sil_score = silhouette_score(X_scaled[mask], labels[mask])
    else:
        sil_score = -1

    mlflow.log_param("algorithm", "DBSCAN")
    mlflow.log_param("eps", eps)
    mlflow.log_param("min_samples", min_samples)

    mlflow.log_metric("silhouette_score", sil_score)
```

```python
# PCA Metrics Tracking
```

```python
from sklearn.decomposition import PCA

with mlflow.start_run(run_name="PCA_Dimensionality"):

    pca = PCA(n_components=3)
    X_pca = pca.fit_transform(X_scaled)

    variance = sum(pca.explained_variance_ratio_)

    mlflow.log_param("n_components", 3)
    mlflow.log_metric("explained_variance", variance)

    print("Explained Variance:", variance)
```
